```python
def scraper(keyword,locations):
    #import the necessory libraries
    import undetected_chromedriver as uc
    from selenium.webdriver.common.by import By
    from selenium.webdriver.common.keys import Keys
    import time
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    driver = uc.Chrome()

    # Open Glassdoor
    driver.get('https://www.glassdoor.com/profile/login_input.htm')
    time.sleep(4)
    driver.maximize_window
    
    # Enter the Gmail to login
    email=driver.find_element(By.ID,'inlineUserEmail')
    email.send_keys("enter ypur mail" + Keys.ENTER) 
    time.sleep(3)
    
    #enter password to login
    password=driver.find_element(By.ID ,'inlineUserPassword')
    password.send_keys('enter yoyr pssword' + Keys.ENTER)
    time.sleep(10)
        
    #click the search button
    find_job = driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/div/div[1]/div/div[3]/div[1]/button')
    find_job.click()
    time.sleep(5)

    #enter keyword of job
    search_job=driver.find_element(By.ID,'sc.keyword')
    search_job.send_keys(keyword)
    time.sleep(4)
  
    #enter location
    location=driver.find_element(By.ID,'sc.location')
    location.send_keys(locations + Keys.ENTER)
    time.sleep(4)

    #to remove POP_UP
    try:
        # Wait for the pop-up to appear
        wait = WebDriverWait(driver, 10)  # Timeout after 10 seconds
        popup_close_button = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//button[@aria-label='Cancel']")
        ))

        # Close the pop-up
        popup_close_button.click()
        print("Pop-up closed successfully.")

        # Add further actions here if needed, e.g., navigating or interacting with the page

    except Exception as e:
        print(f"An error occurred: {e}")
    time.sleep(4)  
    
    #FOR SHOW MORE JOBS
    try:
        while True:
            try:
                # Locate the "Show more jobs" button
                show_more_button = driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/div/div/button")

                
                show_more_button.click()
                print("Clicked 'Show more jobs' button.")

                # Wait for new jobs to load
                time.sleep(2)

            except Exception as e:
                # Break if the button no longer exists or is not clickable
                print("No more 'Show more jobs' button found. All jobs are loaded.")
                break

        # At this point, all jobs should be loaded. Add scraping logic if needed.

    except Exception as e:
        print(f"An error occurred: {e}")

    time.sleep(4)

    import time
    #to check that how many jobs in a wepage
    divs=driver.find_elements(By.CLASS_NAME,'JobsList_jobListItem__wjTHv ')
    div_count=len(divs)
    div_count
    company=[]
    job_title=[]
    location=[]
    link=[]
    time=[]
    for i in range(1,div_count+1):
        #scrap the name of a company that upload a job
        try:
            company.append(driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[' +str(i)+ ']/div/div/div[1]/div[1]/div[1]/div/div/span').text)
        except:
            company.append("#N/A")
            pass
        
        #scrap the title of a job
        try:
            job_title.append(driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[' +str(i)+ ']/div/div/div[1]/div[1]/a[1]').text)
        except:
            job_title.append("#N/A")
            pass
        
        
        #scrap the location of job

        try:
            location.append(driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[' +str(i)+ ']/div/div/div[1]/div[1]/div[2]').text)
        except:
            location.append("#N/A")
            pass
            
        #scrap the link of a job
        try:
            element=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[' +str(i)+ ']/div/div/div[1]/div[1]/a[2]')
            href_element=element.get_attribute('href')
            link.append(href_element)
        except:
            link.append("#N/A")
            pass
            
        #scrap the upload time of job
        try:
            element=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[' +str(i)+ ']/div/div/div[1]/div[2]').text
            time.append(element)
        except:
            time.append("#N/A")
            pass
    #quit the driver    
    driver.quit()
    #return 
    return company,job_title,link,time,location
    
```

```python
keyword = ["Mechanical Engineer Jobs", "Civil Engineer Jobs", "Electrical Engineer Jobs", "Software Engineer Jobs", "Structural Engineer Jobs"]

loca=['Lahore','Karachi','Islamabad']
company=[]
job_title=[]
link=[]
time=[]
location=[]

#nested loop to scrap jobs of different keyword with different location
for i in range(len(keyword)):
    for j in range(len(loca)):
        print('For:',keyword[i],'& ',loca[j])
        company_name,job_name,links,timing,locations=scraper(keyword[i],loca[j])
        company.append(company_name)
        job_title.append(job_name)
        link.append(links)
        time.append(timing)
        location.append(locations)
        print()
    
```

```python
import itertools
```

```python
#convert 2D list to 1D list.
companies = list(itertools.chain.from_iterable(company))
job = list(itertools.chain.from_iterable(job_title))
linkk = list(itertools.chain.from_iterable(link))
place= list(itertools.chain.from_iterable(location))
tim= list(itertools.chain.from_iterable(time))
```
